Cele mai afectate coloane de valorile lipsă sunt acelea pentru care isna().sum() arată cele mai mari valori și care apar mai evidențiate în heatmap. Dacă aceste coloane nu sunt folosite direct pentru antrenarea modelului sau conțin doar informații auxiliare, ele nu sunt foarte importante pentru model și pot fi eliminate fără probleme. Dacă însă lipsurile apar în coloane esențiale, cum ar fi textul recenziei sau eticheta țintă, atunci rândurile respective trebuie eliminate, deoarece modelul nu poate fi antrenat corect fără aceste informații. Pentru alte coloane mai puțin critice, valorile lipsă pot fi completate cu o valoare implicită, cu media, moda sau un text precum „unknown”, în funcție de tipul datelor.

```python
import pandas as pd

# load dataset from GitHub
url = "https://raw.githubusercontent.com/vladimir-dresevic/ml-product-reviews-project/main/data/product_reviews_full.csv"

df = pd.read_csv(url)

# Count the number of rows before removing missing values
rows_before = len(df)

# Filter out rows that contain at least one missing value
rows_with_nan = df[df.isnull().any(axis=1)]

# Display a random sample of rows that will be removed
print("Randomly selected rows containing missing values:\n")
print(rows_with_nan.sample(n=min(5, len(rows_with_nan)), random_state=42))

# Remove rows with any missing values
df_cleaned = df.dropna()

# Count the number of rows after removing missing values
rows_after = len(df_cleaned)

# Show removal statistics
print("\n Removal statistics:")
print(f"- Number of rows before: {rows_before}")
print(f"- Number of rows after: {rows_after}")
print(f"- Number of removed rows: {rows_before - rows_after}")
```

```python
# Step 1: Convert to string and remove the 'USD' prefix and any leading/trailing spaces
df['product_price_cleaned'] = (
    df['product_price']
    .astype(str)
    .str.replace(r'$', '', regex=True)    # Remove '$'
    .str.replace(r'[^\d.]', '', regex=True) # Remove all non-numeric characters except the dot
    .str.strip()
)

# Step 2: Convert cleaned string to float
df['product_price'] = pd.to_numeric(df['product_price_cleaned'], errors='coerce')

# Step 3: Drop the temporary column
df = df.drop(columns=['product_price_cleaned'])

# Step 4: Drop any rows where conversion failed (still NaN)
df = df.dropna(subset=['product_price'])

# Step 5: Confirm result
print("Column type after parsing:", df['product_price'].dtype)
print("\nPrice summary:")
print(df['product_price'].describe())
```

```python
# Step 1: Convert all sentiment values to lowercase and strip extra spaces
df['sentiment'] = df['sentiment'].astype(str).str.lower().str.strip()

# Step 2: Show all unique values in the sentiment column
print("Unique sentiment values after cleaning:")
print(df['sentiment'].value_counts())

# Step 3: Convert column type to 'category'
df['sentiment'] = df['sentiment'].astype('category')
print("\nSentiment column converted to type:", df['sentiment'].dtype)
```

```python
# Drop columns that are not useful for modeling (keeping product_name for categorization)
df = df.drop(columns=['review_uuid'])

# Preview remaining columns
print("Remaining columns:")
print(df.columns.tolist())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Show summary statistics grouped by sentiment
print("Price summary by sentiment:")
print(df.groupby('sentiment', observed=False)['product_price'].describe())

# Boxplot of prices by sentiment
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='sentiment', y='product_price')
plt.title("Product Price Distribution by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Product Price")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

# 1. Re-load the dataset
url = "https://raw.githubusercontent.com/iftode/-ml-product-reviews-project/main/data/IMLP4_13-product_reviews_full.csv"
df_plot = pd.read_csv(url)

# 2. Clean product_price
df_plot['product_price'] = pd.to_numeric(
    df_plot['product_price'].astype(str).str.replace('$', '', regex=False).str.replace(r'[^\d.]', '', regex=True),
    errors='coerce'
)

# 3. Clean sentiment and names
df_plot['sentiment'] = df_plot['sentiment'].astype(str).str.lower().str.strip()
# Remove non-ASCII characters from product names to avoid font errors
df_plot['product_name'] = df_plot['product_name'].apply(lambda x: re.sub(r'[^\x00-\x7f]+', r'', str(x)))

# 4. Filter for Top 10 products for a readable plot
top_products = df_plot['product_name'].value_counts().nlargest(10).index
df_filtered = df_plot[df_plot['product_name'].isin(top_products)].dropna(subset=['product_price', 'sentiment'])

# 5. Generate the plot
plt.figure(figsize=(14, 8))
sns.boxplot(data=df_filtered, x='product_name', y='product_price', hue='sentiment')
plt.title("Product Price Distribution for Top 10 Products by Sentiment")
plt.xlabel("Product Name (Cleaned)")
plt.ylabel("Product Price")
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
